```python
import numpy as np
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

REGRESSION MODEL<br>
- I chose "ADAM" as the optimizer and "RELU" as activation as its the standard for basic models.
- ADAM tries to minimize loss and RELU avoids linearization.
- No activation for the output layer and only one neuron as its regression.
- The mean-squared-error "0.9", so the model predictions are usually "~0.9" off as it is the square root of "0.9.

```python
# CREATING SYNTHETIC DATASET
syn1_x = np.random.rand(1500, 4)
syn1_y = 8*syn1_x[:,0] + 11*syn1_x[:,1] + 41*syn1_x[:,2] + 27*syn1_x[:,3] + np.random.randn(1500)

# SPLITTING SYNTHETIC DATASET
X_train1, X_test1, y_train1, y_test1 = train_test_split(
    syn1_x,
    syn1_y,
    test_size=0.2,
    random_state=42,
    shuffle=True
)

# DEFINING REGRESSION MODEL
model1 = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation="relu", input_shape=(4,)),
    tf.keras.layers.Dense(8, activation="relu", input_shape=(4,)),
    tf.keras.layers.Dense(1)
])

# BUILDING REGRESSION MODEL
model1.compile(
    optimizer="adam",
    loss="mse",
    metrics=['mean_squared_error']
)

# TRAINING REGRESSION MODEL
model1.fit(X_train1, y_train1, epochs=100, batch_size=32, verbose=0)
eval_metric1 = model1.evaluate(X_test1, y_test1)

# EVAL REGRESSION MODEL
print("The Regression Model's Evaluation: " + str(eval_metric1))
```
